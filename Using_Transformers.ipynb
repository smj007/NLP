Sentiment

```
from transformers import pipeline
classifier = pipeline('sentiment-analysis')
classifier('Laurel and Hardy are Greatly Missed')
```

Q&A

```
from transformers import pipeline
question_answerer = pipeline('question-answering')
question_answerer({
  'question': 'What is the name of my dog?',
  'context': 'I have a dog named Sai. He eats everyones share of food.'})
```

Translation

```
from transformers import pipeline
translator = pipeline('translation_en_to_fr')
translator("The quick brown fox jumped.")
```

Summarisation

```
from transformers import pipeline
summarizer = pipeline('summarization', model="t5-base", tokenizer="t5-base", framework="tf")
speech = open('./text1.txt').read()
summarizer(speech, min_length=50, max_length=100)
```
